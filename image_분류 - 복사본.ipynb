```python
temp = []

for i in range(len(excel['내용'])):
    url = excel['내용'][i]
    component = re.findall(r'src="[^"]*\/([^"]*\.(?:jpg|JPG|png|PNG|pdf|PDF|jpeg|JPEG|JFIF|jfif))"',url)
    if component != []:
        temp.append(component)
        
print(temp)
print(len(temp))

image_path = "C:\\Users\\User\\Desktop\\webdata\\sw\\uploads\\image\\"
dest_path = "C:\\Users\\User\\Desktop\\kor-공지사항학부\\"
os.makedirs(dest_path)
l = os.listdir(image_path)
# print(l)


# for i in range(len(temp)):
#     for j in range(len(temp[i])):
# #         print(temp[i][j])

num = 0
for i in range(len(temp)):
    for j in range(len(temp[i])): 
        image = image_path + '\\' + temp[i][j]
        dest = dest_path + '\\' + temp[i][j] 
        target = temp[i][j]
        print(image)
        print(dest)
        if target in l:
            print('ok')
            num = num + 1 
            shutil.copy(image,dest)
        else:
            print('failed')
            
print(num)
```
